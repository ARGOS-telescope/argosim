```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True
```

```python
backends = [r'$\texttt{NumPy}$', r'$\texttt{JAX}$-CPU', r'$\texttt{JAX}$-GPU']
time_tracking_4096 = np.array([13.392, 2.683, 0.008])
time_observation_4096 = np.array([88.799, 4.490, 0.645])
time_total_4096 = time_tracking_4096 + time_observation_4096

time_tracking_1024 = np.array([13.714, 1.002, 0.007])
time_observation_1024 = np.array([87.645, 2.984, 0.098])
time_total_1024 = time_tracking_1024 + time_observation_1024

time_tracking_256 = np.array([13.818, 0.907, 0.008])
time_observation_256 = np.array([83.312, 2.854, 0.086])
time_total_256 = time_tracking_256 + time_observation_256

```

```python
time_tracking_256_bands = np.array([2.121, 0.181, 0.007])
time_observation_256_bands = np.array([23.408, 0.888, 0.026])
time_total_256_bands = time_tracking_256_bands + time_observation_256_bands

time_tracking_512_bands = np.array([4.354, 0.349, 0.007])
time_observation_512_bands = np.array([45.446, 1.650, 0.037])
time_total_512_bands = time_tracking_512_bands + time_observation_512_bands

time_tracking_1024_bands = np.array([13.392, 2.683, 0.007])
time_observation_1024_bands = np.array([88.799, 4.490, 0.062])
time_total_1024_bands = time_tracking_1024_bands + time_observation_1024_bands
```

```python
print(time_total_256_bands)
print(time_total_512_bands)
print(time_total_1024_bands)
```

```python

# Plot the total time bar chart
plt.figure(figsize=(10, 6))

plt.bar(np.arange(3) + 0, time_total_256_bands, color=['deepskyblue', 'teal', 'mediumorchid'])

plt.bar(np.arange(3) + 4., time_total_512_bands, color=['deepskyblue', 'teal', 'mediumorchid'], label=backends)

plt.bar(np.arange(3) + 8, time_total_1024_bands, color=['deepskyblue', 'teal', 'mediumorchid'])



plt.legend()
plt.xticks([1,5,9], [256, 512, 1024])
# plt.title('Total Time for Tracking and Observation')
plt.xlabel('Number of frequency bands')
plt.ylabel('Time (secs)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.yscale('log')
# plt.savefig('total_time_bar_chart.png')
plt.show()
```
